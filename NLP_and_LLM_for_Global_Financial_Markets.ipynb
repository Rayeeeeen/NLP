```python
#to scrap normal pages

from bs4 import BeautifulSoup
import requests
import pandas as pd
from datetime import datetime
#content = page.text
#soup= BeautifulSoup(page.content,"html.parser")
#print(soup.prettify())

#list1=soup.find_all('div', id_="g_1_0jVMbFK5")
def urls(n):# n: number of pages
    url='https://www.boursorama.com/actualite-economique/page-'
    liensutiles= []

    for i in range (1,n+1):


        page = requests.get(url+str(i))
        page = page.text
        soup= BeautifulSoup(page,"html.parser")
        #print(soup.prettify())
        boxes=soup.find_all('div', class_='c-list-details-news__title')
        #print(boxes)
        for box in boxes:
            lien=box.find('a')['href']
            if '/actualite-economique/actualites/' in lien:
                liensutiles.append('https://www.boursorama.com'+lien)


    #print(liensutiles)
    return liensutiles



def data_extraction(url):
    date_format = '%d/%m/%Y à %H:%M'
    liensutiles= []
    page = requests.get(url)
    page = page.text
    soup= BeautifulSoup(page,"html.parser")
    date= soup.find('span', class_='c-source__time').get_text()
    date = datetime.strptime(date, date_format)
    #print(date)
    title= soup.find('h1', class_='c-title-big c-title--color-dark')
    title = title.find('span').get_text()
    body= soup.find ('article', class_='c-news-detail__content')
    body= body.find_all('p')
    body1=''
    for b in body:
        body1+=b.get_text()

    body= body1.strip()
    #print(body)
    return date,title,body
#data_extraction('https://www.boursorama.com/actualite-economique/actualites/des-dizaines-de-morts-dans-les-raids-israeliens-nouvelles-aides-pour-gaza-b723301bb25b96795bc2b119587abdd6')
# print(urls (2))


def table_maker(n):#n : number of pages


    data = {'date': [], 'title': [], 'body':[], }
    df = pd.DataFrame(data)

    liens= urls(n)
    for lien in liens:

        row= data_extraction(lien)
        df.loc[len(df)] = row

    #print(df)
    return df
df= table_maker(2)
df.to_excel('output.xlsx')
```
